## P2 – Distribución Uniforme en $N$ dimensiones

### P2.0 – Descripción teórica

#### Distribución uniforme en una dimensión

La función de densidad de probabilidad (fdp) de una variable continua $X \sim U(a, b)$ es:

$$
f_X(x) = 
\begin{cases}
\frac{1}{b - a}, & \text{si } x \in ]a, b] \\
0, & \text{en otro caso}
\end{cases}
$$

#### Distribución uniforme en $\mathbb{R}^n$

Para un vector aleatorio $X = (X_1, X_2, \dots, X_n)$, donde cada $X_i \sim U(a_i, b_i)$ (independientes), la función de densidad conjunta es:

$$
f_X(\underbar{x}) = f_X(x_{1}, x_{2}, x_{3}, \ldots, x_{n}) = f_{X_{1}}(x_{1}) \cdot f_{X_{2}}(x_{2}) \cdot f_{X_{3}}(x_{3}) \ldots f_{X_{n}}(x_{n})
\\
$$
Por lo tanto:
$$
f_X(x) =
\begin{cases}
\frac{1}{\prod_{i=1}^n (b_i - a_i)}, & \text{si } x_i \in ]a_i, b_i] \ \forall i \\
0, & \text{en otro caso}
\end{cases}
$$

Siendo $\frac{1}{\prod_{i=1}^n (b_i - a_i)}$ la constante de normalización.

### P2.1 – Implementación del método inverso

Para muestrear esta distribución se utiliza el Método Inverso, el cual consiste en generar una realización de una Distribución Uniforme $u$ $\in  ]0,1]$ y luego evaluarla en la función inversa de la función de desisidad de probabilidad acumulada ${F_{X}(\underbar{x})}^{-1}$. Como se tiene que las variables aleatorias $X_{i}$ son idependientes, se aplica el método a cada una de ellas a partir de las siguientes expresiones:
$$
u_{i} = F_{X_{i}}(x_i)
$$
Para una distribución uniforme en una dimensión:
$$
F_{X_{i}}(x_i) = \frac{x_{i} - a_{i}}{b_{i} - a_{i}}
$$
Por lo tanto:
$$
x_{i} = a_{i} + u_{i} \cdot (b_{i} - a_{i})
$$
Siendo $x_{i}$ una muestra de $f_{X_{i}}$.

La distribución uniforme en $n$ dimensiones fue implementada en el archivo `GF7013/probability_functions/pdf/pdf_uniform_nD.py`, dentro del método `_draw`.

### Fragmento de código modificado – Método Inverso (`pdf_uniform_n.py`)

```python
def _draw(self, Ns)
    if Ns is None:
        u = self.rng.uniform(0, 1, size=self.N)
        sample = self.ll + (self.ul - self.ll) * u
    else:
        u = self.rng.uniform(0, 1, size=(self.N, Ns))
        sample = self.ll[:, None] + (self.ul - self.ll)[:, None] * u
    return sample

## Cierre de P2 e introducción a P3 – De lo uniforme a lo normal multivariado

El método inverso, utilizado en la P2, es uno de los mecanismos más directos para generar variables aleatorias cuando se conoce la función de distribución acumulada (CDF). En el caso de una distribución uniforme, el procedimiento es simple y efectivo:

$$
x = a + (b - a) \cdot u, \quad u \sim \mathcal{U}(0, 1)
$$

Este enfoque fue aplicado de forma independiente en cada dimensión, permitiendo generar vectores aleatorios en $\mathbb{R}^2$ con componentes uniformemente distribuidos. El análisis gráfico en la P2.3 mostró cómo, a medida que aumenta el número de muestras ($N = 10^4$, $10^5$, $10^6$), los histogramas se estabilizan y reflejan fielmente la densidad uniforme teórica.

---

Con esto cerramos el estudio de distribuciones con variables independientes y pasamos a un caso más general: la **distribución normal multivariada**, donde las variables están correlacionadas.

En esta parte, el objetivo será generar muestras que sigan una distribución $X \sim \mathcal{N}(\mu, C)$ en $\mathbb{R}^2$, utilizando el método relacional. Este método transforma variables normales estándar independientes en muestras que respetan una media y covarianza dadas:

$$
X = \mu + WZ, \quad Z \sim \mathcal{N}(0, I), \quad C = WW^\top
$$

A continuación, se muestra la implementación de este muestreo y su validación estadística y visual.

### P3.0 – Teoría de la distribución normal multivariada

Una variable aleatoria $X \in \mathbb{R}^n$ sigue una **distribución normal multivariada** si:

$$
X \sim \mathcal{N}(\mu, C)
$$

donde:
- $\mu \in \mathbb{R}^n$ es el **media multivariada**,
- $C \in \mathbb{R}^{n \times n}$ es la **matriz de covarianza** (simétrica y definida positiva).

#### Función de densidad de probabilidad (pdf):

$$
f_X(x) = \frac{1}{\sqrt{(2\pi)^n \det(C)}} \exp\left( -\frac{1}{2}(x - \mu)^\top C^{-1} (x - \mu) \right)
$$

Las marginales de esta distribución son normales univariadas, y si $C$ tiene elementos no nulos fuera de la diagonal, existe correlación entre las variables.

La implementación de esta distribución se realizó en `GF7013/probability_functions/pdf/pdf_normal.py`.

### P3.1 – Método relacional y modificaciones en el código

Para generar muestras desde una distribución normal multivariada se utilizó el **método relacional**:

$$
X = \mu + WZ, \quad Z \sim \mathcal{N}(0, I), \quad C = WW^\top
$$

Donde $W$ es la **descomposición de Cholesky** de la matriz de covarianza $C$.

La modificación principal se realizó en el método `_draw(self, Ns)` del archivo `pdf_normal.py`. A continuación se muestra el código modificado:

```python
def _draw(self, Ns):
    if Ns is None:
        u = self.rng.standard_normal(size=self.N)
        sample = self.mu + self.right_chol_cov @ u
    else:
        u = self.rng.standard_normal(size=(self.N, Ns))
        sample = self.mu[:, np.newaxis] + self.right_chol_cov @ u
    return sample

```python
def _log_likelihood(self, x):
    x = self.__check_x(x)
    misfit = x - self.mu
    return -0.5 * np.dot(misfit.T, np.dot(self.inv_cov, misfit))

def _likelihood(self, x):
    LogLike = self._log_likelihood(x)
    return np.exp(LogLike)

Finalmente, se agregó una función de verificación `test_draw_samples()` que compara la media y covarianza empírica con los valores teóricos definidos:

```python
def test_draw_samples(self, samples, tol_mu=0.05, tol_cov=0.1, verbose=True):
    mu_emp = np.mean(samples, axis=1)
    cov_emp = np.cov(samples)

    error_mu = np.abs(mu_emp - self.mu)
    error_cov = np.abs(cov_emp - self.cov)

    max_err_mu = np.max(error_mu)
    max_err_cov = np.max(error_cov)

    if verbose:
        print(f"Media empírica: {mu_emp}")
        print(f"Error media: {error_mu}")
        print(f"Error covarianza:\n{error_cov}")
        print(f"Max error media: {max_err_mu:.4f} | Tolerancia: {tol_mu}")
        print(f"Max error cov: {max_err_cov:.4f} | Tolerancia: {tol_cov}")

        if max_err_mu < tol_mu and max_err_cov < tol_cov:
            print("✅ TEST PASÓ")
        else:
            print("❌ TEST NO PASÓ")

### P3.2 – Histograma conjunto (visualización bidimensional)

El histograma conjunto muestra cómo se distribuyen las muestras de $X = [X_1, X_2]^\top$ generadas a partir de la distribución normal multivariada con:

- Media: $\mu = [0.5,\ 3.0]$
- Covarianza: $C = \begin{bmatrix} 2 & 1 \\ 1 & 4 \end{bmatrix}$

La forma elíptica de la nube de puntos es coherente con lo que se espera de una normal multivariada. La orientación diagonal de la elipse refleja la correlación positiva entre las variables $X_1$ y $X_2$ (término $C_{12} = 1$).

Además, el centro de la nube está desplazado respecto al origen, ubicado alrededor del vector media $\mu$, como corresponde.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from GF7013.probability_functions import pdf

# Definir parámetros y crear instancia
par = {
    'mu': np.array([0.5, 3.0]),
    'cov': np.array([[2, 1], [1, 4]])
}
normal = pdf.pdf_normal(par)

# Generar muestras
Ns = int(1e4)
samples = normal.draw(Ns)
mu = par['mu']
cov = par['cov']

# ---------- Histograma conjunto ----------
plt.figure(figsize=(6, 5))
plt.hist2d(samples[0], samples[1], bins=50, cmap='viridis')
plt.title(f'Histograma conjunto (Ns = {Ns})')
plt.xlabel('X1')
plt.ylabel('X2')
plt.colorbar(label='Frecuencia')
plt.tight_layout()
plt.show()

# ---------- Histogramas marginales ----------
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

for i in range(2):
    x_vals = samples[i, :]
    axs[i].hist(x_vals, bins=50, density=True, color='lightblue', edgecolor='black', label='Histograma')

    x_line = np.linspace(x_vals.min(), x_vals.max(), 200)
    y_line = norm.pdf(x_line, loc=mu[i], scale=np.sqrt(cov[i, i]))
    axs[i].plot(x_line, y_line, color='red', label='FDP teórica')

    axs[i].set_title(f'Variable X{i+1} – Ns = 1e4')
    axs[i].legend()
    axs[i].grid()

plt.tight_layout()
plt.show()
````
